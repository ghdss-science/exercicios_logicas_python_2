```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Gráfico = Principais Desafios na Carreira Docente (Atualizado) "

```python
# Dados para o gráfico atualizado
categorias_atualizadas = [
    "Falta de valorização e estímulo",
    "Falta de apoio/reconhecimento da sociedade"
]
percentuais_atualizados = [74.8, 61.3]

# Configuração do gráfico atualizado
plt.figure(figsize=(8, 4))
plt.barh(categorias_atualizadas, percentuais_atualizados, color='lightcoral', edgecolor='black')
plt.xlabel("Percentual (%)", fontsize=12)
plt.ylabel("Desafios na Carreira Docente", fontsize=12)
plt.title("Principais Desafios na Carreira Docente (Atualizado)", fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Exibição do gráfico atualizado
plt.show()
```

## Interpretação

*** Falta de valorização e estímulo é o maior desafio, com uma proporção acima de 70%.

*** Falta de apoio/reconhecimento da sociedade também é significativa, mas parece um pouco menor (provavelmente entre 50% e 60%).

### Gráfico 2 = Fatores de Desvalorização da Docência 

```python
# Dados
fatores = ['Econômico', 'Social', 'Psicológico', 'Obsolescência', 'Desqualificação']
impacto = [30, 25, 20, 15, 10]  # Exemplo de pesos (personalizável)

# Configuração do gráfico
plt.figure(figsize=(10, 6))
plt.bar(fatores, impacto, color='purple', edgecolor='black')
plt.xlabel('Fatores', fontsize=12)
plt.ylabel('Impacto (%)', fontsize=12)
plt.title('Fatores de Desvalorização da Docência', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Interpretação 2

*** O fator mais significativo tem impacto em torno de 30%, sendo o mais elevado entre os fatores listados. 

*** Há uma distribuição decrescente nos demais fatores, sugerindo que alguns têm impacto moderado ou baixo. 

*** Este tipo de visualização facilita a análise dos fatores mais críticos, ajudando a identificar áreas prioritárias para intervenção ou discussão.

### Gráfico 3 = Carga de Trabalho: Brasil x Outros Países 

```python
import matplotlib.pyplot as plt

# Dados
categorias = ['Brasil', 'Países com melhores indicadores']
valores = [300, 200]  # Brasil tem carga 50% maior

# Configuração do gráfico
plt.figure(figsize=(8, 5))
plt.bar(categorias, valores, color=['orange', 'green'], edgecolor='black')
plt.xlabel('Região', fontsize=12)
plt.ylabel('Média de Alunos por Professor', fontsize=12)
plt.title('Carga de Trabalho: Brasil x Outros Países', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Interpretação 

*** A média de alunos por professor no Brasil é significativamente maior, indicando uma sobrecarga na carga de trabalho docente em comparação com países de referência.

*** Essa diferença reflete possíveis desafios estruturais no sistema educacional brasileiro, como a alta demanda por professores e insuficiência de recursos.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Gr√°fico = Principais Desafios na Carreira Docente (Atualizado) "

```python
# Dados para o gr√°fico atualizado
categorias_atualizadas = [
    "Falta de valoriza√ß√£o e est√≠mulo",
    "Falta de apoio/reconhecimento da sociedade"
]
percentuais_atualizados = [74.8, 61.3]

# Configura√ß√£o do gr√°fico atualizado
plt.figure(figsize=(8, 4))
plt.barh(categorias_atualizadas, percentuais_atualizados, color='lightcoral', edgecolor='black')
plt.xlabel("Percentual (%)", fontsize=12)
plt.ylabel("Desafios na Carreira Docente", fontsize=12)
plt.title("Principais Desafios na Carreira Docente (Atualizado)", fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Exibi√ß√£o do gr√°fico atualizado
plt.show()
```

## Interpreta√ß√£o

*** Falta de valoriza√ß√£o e est√≠mulo √© o maior desafio, com uma propor√ß√£o acima de 70%.

*** Falta de apoio/reconhecimento da sociedade tamb√©m √© significativa, mas parece um pouco menor (provavelmente entre 50% e 60%).

### Gr√°fico 2 = Fatores de Desvaloriza√ß√£o da Doc√™ncia 

```python
# Dados
fatores = ['Econ√¥mico', 'Social', 'Psicol√≥gico', 'Obsolesc√™ncia', 'Desqualifica√ß√£o']
impacto = [30, 25, 20, 15, 10]  # Exemplo de pesos (personaliz√°vel)

# Configura√ß√£o do gr√°fico
plt.figure(figsize=(10, 6))
plt.bar(fatores, impacto, color='purple', edgecolor='black')
plt.xlabel('Fatores', fontsize=12)
plt.ylabel('Impacto (%)', fontsize=12)
plt.title('Fatores de Desvaloriza√ß√£o da Doc√™ncia', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Gr√°fico 3 = Carga de Trabalho: Brasil x Outros Pa√≠ses 

```python
import matplotlib.pyplot as plt

# Dados
categorias = ['Brasil', 'Pa√≠ses com melhores indicadores']
valores = [300, 200]  # Brasil tem carga 50% maior

# Configura√ß√£o do gr√°fico
plt.figure(figsize=(8, 5))
plt.bar(categorias, valores, color=['orange', 'green'], edgecolor='black')
plt.xlabel('Regi√£o', fontsize=12)
plt.ylabel('M√©dia de Alunos por Professor', fontsize=12)
plt.title('Carga de Trabalho: Brasil x Outros Pa√≠ses', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

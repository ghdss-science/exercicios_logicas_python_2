#### 1. Crie uma classe para representar uma pessoa, com os atributos privados de nome, idade e altura. Crie os métodos públicos necessários para sets e gets e também um método para imprimir os dados de uma pessoa.

```python
class Pessoa: 
    
    def __init__(self): 
        
        self.__nome = "" 
        self.__idade = 0 
        self.__altura = 0.0 
        
    # Getter e Setter para nome 
    def get_nome(self): 
        
        return self.__nome
    
    def set_nome(self,nome): 
        
        self.__nome = nome 
        
    # Getter e Setter para idade 
    def get_idade(self): 
        
        return self.__idade 
    
    def set_idade(self, idade): 
        
        self.__idade = idade 
        
    # Getter e Setter para altura 
    def get_altura(self): 
        
        return self.__altura 
    
    def set_altura(self, altura): 
        
        self.__altura = altura 
        
    # Método para imprimir os dados
    def imprimir(self): 
        
        print(f"Nome: {self.get_nome()} | Idade: {self.get_idade()} | Altura: {self.get_altura()}")
```

#### 2. Crie uma classe Agenda que pode armazenar 10 pessoas e seja capaz de realizar as seguintes operações:

void armazenaPessoa(String nome, int idade, float altura);

void removePessoa(String nome);

int buscaPessoa(String nome); // informa em que posição da agenda está a pessoa

void imprimeAgenda(); // imprime os dados de todas as pessoas da agenda

void imprimePessoa(int index); // imprime os dados da pessoa que está na posição “i” da agenda



```python
class Pessoa: 
    
    def __init__(self, nome: str, idade: int, altura: float): 
        
        self.nome = nome 
        self.idade = idade
        self.altura = altura 
        
    def imprimir(self): 
        
        print(f"Nome: {self.nome} | Idade: {self.idade} | Altura: {self.altura: .2f}m") 
```

```python
class Agenda: 
    
    def __init__(self): 
        
        self.pessoas = [] 
        
    def armazena_pessoa(self, nome: str, idade: int, altura: float): 
        
        if len(self.pessoas) < 10: 
            
            self.pessoas.append(Pessoa(nome, idade, altura)) 
            print("Pessoa adicionada com sucesso") 
            
        else: 
            
            print("Agenda cheia ! Não é possível adicionar mais pessoas. ") 
            
    def remove_pessoa(self, nome: str): 
        
        for i, pessoa in enumerate(self.pessoas): 
            
            if pessoa.nome.lower() == nome.lower(): 
                
                del self.pessoas[i]
                print(f"{nome} removido com sucesso!") 
                return 
            
        print(f"Pessoa com nome '{nome}' não encontrado. ") 
        
    def busca_pessoa(self, nome: str): 
        
        for i, pessoa in enumerate(self.pessoas): 
            
            if pessoa.nome.lower() == nome.lower(): 
                
                return i 
            
        return -1
    
    def imprime_agenda(self): 
        
        if not self.pessoas: 
            
            print("Agenda vazia.") 
            
        else: 
            
            for i, pessoa in enumerate(self.pessoas): 
                
                print(f"[{i}]", end="") 
                pessoa.imprimir() 
                
    def imprime_pessoa(self, index: int): 
        
        if 0 <= index < len(self.pessoas): 
            
            self.pessoas[index].imprimir() 
            
        else: 
            
            print("Índice inválido.") 
```

```python
def main(): 
    
    agenda = Agenda() 
    
    while True: 
        
        print("\n --- Menu ---")
        print("1 - Armazenar Pessoa") 
        print("2 - Remover Pessoa") 
        print("3 - Buscar Pessoa") 
        print("4 - Imprimir Agenda") 
        print("5 - Imprimir Pessoa por posição") 
        print("6 - Sair") 
        
        try: 
            
            opcao = int(input("Escolha uma opção: "))
            
        except ValueError: 
            
            print("Por Favor, insira um número válido.") 
            continue 
            
        if opcao == 1: 
            
            nome = input("Nome: ") 
            idade = int(input("Idade: ")) 
            altura = float(input("Altura (em metros): ")) 
            agenda.armazena_pessoa(nome, idade, altura) 
            
        elif opcao == 2: 
            
            nome = input("Nome da pessoa a remover: ") 
            agenda.remove_pessoa(nome) 
            
        elif opcao == 3: 
            
            nome = input("Nome da pessoa a buscar: ") 
            pos = agenda.busca_pessoa(nome) 
            
            if pos != -1: 
                
                print(f"{nome} está na posiçao {pos}. ") 
                
            else: 
                
                print("Pessoa não encontrada.") 
                
        elif opcao == 4: 
            
            agenda.imprime_agenda() 
            
        elif opcao == 5: 
            
            index = int(input("Informe o índice da pessoa: ")) 
            agenda.imprime_pessoa(index) 
            
        elif opcao == 6: 
            
            print("Saindo...") 
            break 
            
        else: 
            print("Opção inválida!") 
```

```python
if __name__ == "__main__":
    
    main()
```

#### 3. Crie uma classe denominada Elevador para armazenar as informações de um elevador dentro de um prédio. A classe deve armazenar o andar atual (térreo = 0), total de andares no prédio, excluindo o térreo, capacidade do elevador, e quantas pessoas estão presentes nele.

#### A classe deve também disponibilizar os seguintes métodos:

Inicializa: que deve receber como parâmetros a capacidade do elevador e o total de andares no prédio (os elevadores sempre começam no térreo e vazio);

Entra: para acrescentar uma pessoa no elevador (só deve acrescentar se ainda houver espaço);

Sai: para remover uma pessoa do elevador (só deve remover se houver alguém dentro dele);

Sobe: para subir um andar (não deve subir se já estiver no último andar);

Desce: para descer um andar (não deve descer se já estiver no térreo);

Encapsular todos os atributos da classe (criar os métodos set e get).



```python
class Elevador: 
    
    def __init__(self): 
        
        self._andar_atual = 0
        self._total_andares = 0
        self._capacidade = 0 
        self._pessoas_presentes = 0
        
    def inicializa(self, capacidade, total_andares): 
        
        self._capacidade = capacidade 
        self._total_andares = total_andares 
        self._andar_atual = 0 
        self._pessoas_presentes = 0 
        
    def entra(self): 
        
        if self._pessoas_presentes < self._capacidade: 
            
            self._pessoas_presentes += 1
            print("Uma pessoa entrou no elevador.") 
            
        else: 
            
            print("Elevador cheio !") 
            
    def sai(self): 
        
        if self._pessoas_presentes > 0: 
            
            self._pessoas_presentes -= 1 
            print("Uma pessoa saiu do elevador.") 
            
        else: 
            
            print("O elevador está vazio !")  
            
    def sobe(self): 
        
        if self._andar_atual < self._total_andares: 
            
            self._andar_atual += 1
            print(f"O elevador subiu para o andar {self._andar_atual}.") 
            
        else: 
            
            print("O elevador já está no último andar.") 
            
    def desce(self): 
        
        if self._andar_atual > 0: 
            
            self._andar_atual -= 1
            print(f"O elevador desceu para o andar {self._andar_atual}.") 
            
        else: 
            
            print("O elevador já está no térreo.") 
            
    # Métodos getters (acesso aos atributos encapsulados) 
    def get_andar_atual(self): 
        
        return self._andar_atual 
    
    def get_total_andares(self): 
        
        return self._total_andares 
    
    def get_capacidade(self): 
        
        return self._capacidade 
    
    def get_pessoas_presentes(self): 
        
        return self._pessoas_presentes
```

```python
def main(): 
    
    elevador = Elevador() 
    
    capacidade = int(input("Informe a capacidade do elevador: ")) 
    total_andares = int(input("Informe o total de andares (sem contar o térreo): ")) 
    elevador.inicializa(capacidade, total_andares) 
    
    while True: 
        
        print("\n---Menu---") 
        print("1 - Entrar uma pessoa") 
        print("2 - Sair uma pessoa") 
        print("3 - Subir uma pessoa") 
        print("4 - Descer uma pessoa") 
        print("5 - Mostrar uma pessoa") 
        print("6 - Mostrar uma pessoas") 
        print("7 - Sair") 
        
        opcao = input("Escolha uma opção: ") 
        
        if opcao == "1": 
            
            elevador.entra() 
            
        elif opcao == "2": 
            
            elevador.sai() 
            
        elif opcao == "3": 
            
            elevador.sobe() 
            
        elif opcao == "4": 
            
            elevador.desce() 
            
        elif opcao == "5": 
            
            print(f"Andar atual: {elevador.get_andar_atual()}") 
            
        elif opcao == "6": 
            
            print(f"Pessoas presentes: {elevador.get_pessoas_presentes()}") 
            
        elif opcao == "7": 
            
            print("Programa encerrado.") 
            break 
            
        else: 
            
            print("Opção inválida. Tente novamente.") 
            
if __name__ == "__main__": 
    
    main()
```

#### 4. Crie uma classe Televisão e uma classe ControleRemoto que pode controlar o volume e trocar os canais da televisão.

O controle de volume permite aumentar ou diminuir a potência do volume de som em uma unidade de cada vez;

O controle de canal também permite aumentar e diminuir o número do canal em uma unidade, porém, também possibilita trocar para um canal indicado;

Também devem existir métodos para consultar o valor do volume de som e o canal selecionado.



```python
class Televisao: 
    
    def __init__(self, canal = 1, volume = 10): 
        
        self._canal = self._validar_canal(canal) 
        self._volume = volume if 0 <= volume <= 50 else 0 
        
    def _validar_canal(self, canal): 
        
        # Canal deve ser entre 1 e 500 e múltiplo de 3 ou 5 
        if canal < 1 or canal > 500: 
            
            return 1 
        
        for i in range(canal, 501): 
            
            if i % 3 == 0 or i % 5 == 0: 
                
                return i 
            
            return 1 
        
    def add_volume(self): 
        
        if self._volume < 50: 
            
            self._volume += 1 
            
    def sub_volume(self): 
        
        if self._volume > 0: 
            
            self._volume -= 1 
            
    def add_canal(self): 
        
        self._canal = self._validar_canal(self._canal + 1) 
        
    def sub_canal(self): 
        
        novo_canal = self._validar_canal(self._canal - 1) 
        if novo_canal >= 1:
            
            self._canal = novo_canal
            
    def set_canal(self, canal): 
        
        self._canal = self._validar_canal(canal) 
        
    def get_canal(self): 
        
        return self._canal 
    
    def get_volume(self): 
        
        return self._volume 
```

```python
class ControleRemoto: 
    
    def __init__(self): 
        
        self.televisao = Televisao() 
        
    def aumentar_volume(self): 
        
        self.televisao.add_volume() 
        
    def diminuir_volume(self): 
        
        self.televisao.sub_volume() 
        
    def aumentar_canal(self): 
        
        self.televisao.add_canal()
        
    def diminuir_canal(self): 
        
        self.televisao.sub_canal() 
        
    def trocar_canal(self, canal): 
        
        self.televisao.set_canal(canal) 
        
    def consultar_canal(self): 
        
        return self.televisao.get_canal() 
    
    def consultar_volume(self): 
        
        return self.televisao.get_volume() 
```

```python
# Simulação de uso 
controle = ControleRemoto() 

acoes = [
    
    ("Aumentar Canal", controle.aumentar_canal), 
    ("Aumentar Canal", controle.aumentar_canal), 
    ("Aumentar Canal", controle.aumentar_canal), 
    ("Aumentar Volume", controle.aumentar_volume), 
    ("Aumentar Volume", controle.aumentar_volume), 
    ("Aumentar Volume", controle.aumentar_volume), 
    ("Trocar Canal para 400", lambda: controle.trocar_canal(400)), 
    ("Aumentar Canal", controle.aumentar_canal), 
    ("Aumentar Canal", controle.aumentar_canal), 
]
```
